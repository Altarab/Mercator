```python
import pandas as pd
import numpy as np
import os
import sys

df = pd.read_csv('insert_coord_mineral.csv', delimiter = ',',encoding='utf-8')
#print(df)


mine = []
f=open('insert_coord.sql',"w+",encoding='utf8')
df = df.reset_index()  # make sure indexes pair with number of rows

for index, row in df.iterrows():
    if row['mine'] not in mine :
        f.write('UPDATE public.mctor_locations SET latitude = '+str(row['latitude'])+', longitude = '+str(row['longitude'])+' WHERE level_value = \''+str(row['mine'])+'\';\n')
f.close()
```
